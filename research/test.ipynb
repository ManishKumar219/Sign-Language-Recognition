```python
X = df_encoded.drop('Label_Encoded', axis = 1)
y = df_encoded[['Label_Encoded']]
print("Landmarks Data of Images:")
display(X.head())
print("\nLables of Landmarks: ")
display(y.head())
```

```python
print(df_encoded.isnull().sum().sum())
pd.DataFrame((df_encoded.isnull().sum()), columns = ['Sum of Null Values'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print([X_train.shape, y_train.shape], [X_test.shape, y_test.shape])
input_shape = X_train.shape[1]
input_shape
```

```python
model = Sequential()

model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(input_shape,1)))
model.add(Dropout(0.2))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Dropout(0.5))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(39, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=100, validation_split = 0.2)
```

```python
accuracyReport = model.evaluate(X_test, y_test)

print("Accuracy on test data: "+ str(round(accuracyReport[1] * 100, 2)) +"%")
```
